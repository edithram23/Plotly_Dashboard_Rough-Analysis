```python
country_mapping = {
    'Congo, Dem. Rep.': 'Congo, The Democratic Republic of the',
    'Congo, Rep.': 'Congo',
    "Cote d'Ivoire": 'Côte d\'Ivoire',
    'Hong Kong, China': 'Hong Kong',
    'Iran': 'Iran, Islamic Republic of',
    'Korea, Dem. Rep.': 'Korea, Democratic People\'s Republic of',
    'Korea, Rep.': 'Korea, Republic of',
    'Reunion': 'Réunion',
    'Swaziland': 'Eswatini',
    'Syria': 'Syrian Arab Republic',
    'West Bank and Gaza': 'Palestine, State of',
    'Yemen, Rep.': 'Yemen',
}

valid_country_names = []
invalid_country_names = []

for country_name in country:
    if country_name in country_mapping:
        valid_country_names.append(pycountry.countries.search_fuzzy(country_mapping[country_name])[0].alpha_3)
    else:
        valid_country_names.append(pycountry.countries.search_fuzzy(country_name)[0].alpha_3)
        
```
