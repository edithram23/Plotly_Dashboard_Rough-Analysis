```python
data=data[data['Name']!='Other']
```

```python
country=data["Name"][1::].unique().tolist()
```

```python
NA_LIST = {
    'United States of America': 'US',
    'Republic of Korea': 'KR',
    'The United Kingdom': 'GB',
    'Türkiye': 'TR',
    'Iran (Islamic Republic of)': 'IR',
    'Bolivia (Plurinational State of)': 'BO',
    'occupied Palestinian territory, including east Jerusalem': 'PS',
    'Republic of Moldova': 'MD',
    'Venezuela (Bolivarian Republic of)': 'VE',
    'Kosovo[1]': 'XK',
    'Democratic Republic of the Congo': 'CD',
    'Côte d’Ivoire': 'CI',
    'United Republic of Tanzania': 'TZ',
    'Micronesia (Federated States of)': 'FM',
    'United States Virgin Islands': 'VI',
    'Northern Mariana Islands (Commonwealth of the)': 'MP',
    'Saint Martin': 'MF',
    'Sint Maarten': 'SX',
    'Bonaire': 'BQ',
    'British Virgin Islands': 'VG',
    'Sint Eustatius': 'BQ',
    'Saba': 'BQ',
    'Holy See': 'VA',
    'Pitcairn Islands': 'PN',
    "Democratic People's Republic of Korea": 'KP'
}
```

```python
ISO=['GLOBAL']
for country_name in country:
    try:
        country = pycountry.countries.get(name=country_name)
        ISO.append(country.alpha_2)
    except Exception as e:
        
        ISO.append(NA_LIST[country_name])
```

```python
ISO
```

```python
data["ISO"] = ISO
```
